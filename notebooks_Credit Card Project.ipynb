#### __2) Analytic approach__

The analytic approach is based on three stages:

1) Descriptive approach - to analyze the expenses distribution - we will use data aggregation to analyze subtotal values and data visualization to see un a graphic way the highest amounts.

2) Diagnostic approach - to see the highest expenses description and analyze if they are correlated a seasonal reason, fixed expenses, ocational expenses , etc.

3) Classification approach - to agrupate expenses and determine the general reason, food, health, transport, etc.
